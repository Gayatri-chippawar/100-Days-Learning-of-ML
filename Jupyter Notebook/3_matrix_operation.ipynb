```python
#importing the numpy array
import numpy as np 
```

```python
#matirx addition 
#the matrix can only be added when there shape is same there dimension should always be same 
#lets take the random integer matrix 
a=np.random.randint(10,20,(3,3))
b=np.random.randint(30,40,(3,3))
print("first matrix is")
print(a)
print("the second matrix is")
print(b)
```

```python
#lets perform the matrix addtion 
print("using function")
print(np.add(a,b))
print("manually without function")
print(a+b)
```

```python
#lets perfrom the matrix subtraction
print("using function")
print(np.subtract(a,b))
print("manually without function")
print(a-b)
```

```python
#matrix with scaler multiplication
print(5*a)
```

```python
#matrix multiplication
print("using funciton")
print(np.multiply(a,b))
print("manually without function")
print(a*b)
```

```python
import numpy as np
```

```python
matrix_1 = np.random.randint(9,19,(2,3))
print("Matrix 1 :")
print(matrix_1)
matrix_2 = np.random.randint(7,21,(2,3))
print("Matrix 2 :")
print(matrix_2)
```

```python
print("Using Function :")
print(np.add(matrix_1 , matrix_2))
print("Manual :")
print(matrix_1 + matrix_2)
```

```python
print("Using Function :")
print(np.multiply(matrix_1 , matrix_2))
print("Manual :")
print(matrix_1 * matrix_2)
```

```python
print("Using Function :")
print(np.subtract(matrix_1 , matrix_2))
print("Manual :")
print(matrix_1 - matrix_2)
```

```python
print("Using Function :")
print(np.divide(matrix_1 , matrix_2))
print("Manual :")
print(matrix_1 / matrix_2)
```

```python
print("Scalor Multiplication :")
division = np.divide(matrix_1,matrix_2)
print(10*division)
```
