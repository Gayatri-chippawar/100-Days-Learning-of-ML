```python
#matrix with all the values 1
mat=np.ones((3,3))
print(mat)
```

```python
#the values are in float let's change it to integer
mat=np.ones((3,3),dtype=int)
print(mat)
```

```python
#all the values with zero
mat=np.zeros((3,3))
print(mat)
```

```python
#now convert the values from float to int 
mat=np.zeros((3,3),dtype=int)
print(mat)
```
