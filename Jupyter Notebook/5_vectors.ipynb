```python
#plotting a vector 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting a vector 
plot_a=plt.quiver(0,0,2,3) #quiver is the function to plot the vectors inside the graphs 
plt.show()
```

```python
#with more and detailed desciption 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#vector substraction 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
#result
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#vector addition 
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
#result
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#operations in vectors 
#dot product 
#vectora ,vectorb a.b=|a||b|costheta
a=np.array([1,2,3])
b=np.array([3,4,5])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

```python
#1*3+2*4+3*5.angle
```

```python
#cross product 
#|a||b|sintheta
```

```python
a=np.array([2,3])
b=np.array([4,4])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
#projection of a on b
#use mathematical formula for projection
a=np.array([1,2])
b=np.array([2,3])
magnitude_of_b=np.sqrt(sum(b**2))
a_dot_b=np.dot(a,b)
proja=(a_dot_b/magnitude_of_b**2)*b
print(proja)
```
