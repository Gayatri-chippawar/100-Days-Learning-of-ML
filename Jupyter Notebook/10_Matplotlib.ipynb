```python
#importing the matplotlib 
import matplotlib.pyplot as plt
```

```python
#importing the numpy
import numpy as np 
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(z)
```

```python
#lets plot the data
#here we are trying to create the sin wave 
plt.plot(x,y)
plt.show()
```

```python
# plotting the cosine curve 
plt.plot(x,z)
plt.show()
```

```python
#adding the details to the curve 
plt.plot(x,y)
plt.title("sine wave")
plt.xlabel('angle')
plt.ylabel("sine value")
plt.show()

          
```

```python
#adding the details to the curve 
plt.plot(x,z)
plt.title("cos wave")
plt.xlabel('angle')
plt.ylabel("cosine value")
plt.show()
```

```python
#drawing the parabola 
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y,'y*')
plt.title("parabola")
plt.show()
```

```python
#plotting the 2 graphs in one plane 
x=np.linspace(-10,5,100)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'y--')
plt.title("sin and cos")
plt.show()
```

```python
#barplot 
fig=plt.figure() #used to make the blank axis 
ax=fig.add_axes([0,0,1,1])
languages=['marathi','hindi','english','spanish']
people=[100,400,50,500]
ax.bar(languages,people)
plt.ylabel('People')
plt.xlabel('languages')
plt.show()
```

```python
#pie chart 
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['marathi','hindi','spanish']
people=[101,450,399]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot 
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y)
ax.scatter(x,z)
plt.show()
```

```python
#3d scatter plot 
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Greens")
plt.show()
```
