```python
#pandas 
#the main work of pandas dataframe is to label the unlabeled data
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
#storing the dataset inside the data1 variable 
data1=fetch_california_housing()
print(data1)
```

```python
#we have used the dataframe to convert the raw data to pandas dataframe
dataset=pd.DataFrame(data1.data,columns=data1.feature_names)
```

```python
#used to print first 5 values 
dataset.head()
```

```python
#tail is used to print last 5 values 
dataset.tail()
```

```python
#no of rows and columns in the dataset 
dataset.shape
```
